## General Python (mostly lists)


```python
# list.count(elmt) -> Returns the number of times that element occurs in the list
x=[1,2,2,3,4]
x.count(2)
```

```python
# list.index(elmt) -> Returns the index of the first occurence of elmt
x=[1,2,3,4,5]
x.index(3)
```

```python
# Sum of lists -> This DOES NOT BEHAVE TERMWISE!!!! It's not like Mathematica! When you "+" two lists, you get a concatenation!
x=[1,2]+[3,4]
print(x)
```

```python
# Question: what is the output of this code?
x=[15,10,2,84]+[1,4,8,7,9]
x.index(x.count(x[0]))
```

```python
# Array Slicing: x[a:b] is element a up to element b-1
x=[1, 2, 3, 4]
print(x[1:3])
```

```python
# Negative indices
x = [0, 1, 2, 3, 4]
print(x[-5:-1])
# Basically you just count backwards. Does not loop, but won't break either. If the end is greater than 0, then it defaults to starting at 0
print(x[-5:2])
```

## Working with Data (pandas)

```python
# It's easier to deal with filtering a dataset if you create the filter as its own variable first!
data = {
    "month" : [1, 1, 2, 3, 4],
    "day" : [1, 2, 2, 3, 4],
    "vibe" : ['good','bad', 'rancid', 'bad','good']
              }
import pandas as pd
df = pd.DataFrame(data)
filters = ((df["month"] != 1) & (df['vibe'] != "rancid"))
df[filters]
```

```python
# .loc[] -> Access a particular column or row of a dataframe(?)
print("Everything in the first row (since no index was set)\n",df.loc[0], "\n")
# .loc[:,"col"] will print the entire column
print("Everything in the \"month\" column \n", df.loc[:,"month"])
```
