Working through video 1 in "Joining Data with Pandas", the video uses Chicago dataportal data (wards) and census data for examples

```python
import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cityofchicago.org", None)

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("htai-wnw4", limit=2000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
results_df.columns
```

Note: Socrata -> government, NGO, non-profit datasets. Check the site for code snippets to import datasets 

```python
# For consistency with videos, turn that "results" dataset into "ward" with columns ward, alderman, address, zip
wards = results_df[['ward','alderman','address','zipcode']]
```

- Looks like that dataset isn't available, or at least I can't find it. I'll grab a different dataset where ward is one of the variables and work from there, since the point is to do an inner-join
- "ACS 5 Year Data by Ward" is interesting at least, so I'll grab that.

```python
# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("awnk-6fvc", limit=2000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)

census = results_df[['ward','total_population']]
```

```python
wards_census=wards.merge(census, on="ward")
wards_census.head(4)
```

```python
# Why do this? Ex. Sort merged dataset by total population, find the name of the Alderman with the most populous ward
wards_census.sort_values('total_population',ascending=False)[['alderman']].head(1)
```
