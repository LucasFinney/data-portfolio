# List Comprehensions
- More efficient way of making a new list based on an old list (ex. add 1 to each value)
- Can write a list comprehension over any iterable
- Components:
    - Iterable
    - Iterator variable (representing members of iterable)
    - Output expression


```python
nums = [4, 8, 15, 16, 23, 42]
# Inefficient way with a for loop:
new_nums = []
for num in nums:
    new_nums.append(num+1)
print(new_nums)

# Efficient way with list comprehension:
new_nums = [num+1 for num in nums]
print(new_nums)
```

```python
# List comprehension with a range() iterable
result = [num**2 for num in range(11)]
print(result)
```

## Nested Loops

```python
# Loop version
pairs_1 = []
for num1 in range(0,2):
    for num2 in range(6,8):
        pairs_1.append((num1,num2))
print(pairs_1)

pairs_2 = [(num1, num2) for num1 in range(0,2) for num2 in range(6,8)]
print(pairs_2)
```
