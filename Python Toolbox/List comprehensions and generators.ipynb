# List Comprehensions
- More efficient way of making a new list based on an old list (ex. add 1 to each value)
- Can write a list comprehension over any iterable
- Components:
    - Iterable
    - Iterator variable (representing members of iterable)
    - Output expression


````python
nums = [4, 8, 15, 16, 23, 42]
# Inefficient way with a for loop:
new_nums = []
for num in nums:
    new_nums.append(num+1)
print(new_nums)

# Efficient way with list comprehension:
new_nums = [num+1 for num in nums]
print(new_nums)
````

````python
# List comprehension with a range() iterable
result = [num**2 for num in range(11)]
print(result)
````

## Nested Loops

````python
# Loop version
pairs_1 = []
for num1 in range(0,2):
    for num2 in range(6,8):
        pairs_1.append((num1,num2))
print(pairs_1)

pairs_2 = [(num1, num2) for num1 in range(0,2) for num2 in range(6,8)]
print(pairs_2)
````

# Advanced Comprehensions

## Conditionals in comprehensions
### List Comprehensions
- Can include "if" statements within a list comprehension
    - ``` [output for iterator in iterable if condition]. ```
    - ``` [output if condition else output_2 for iterator] ```

````python
# Ex. Project every odd number from 0 to 10 to its tenth (ie 1->0.1, 3-> 0.3, etc...), while leaving the even numbers
weird_list = [num/10  if num % 2 != 0 else num for num in range(10)]
print(weird_list)
````

````python
names = ['Alice', 'Bob', 'Billy']
print(list(name for name in names if name[0]=='B'))
````

## Dictionary Comprehensions
- Use curly braces instead of square brackets for a dict comprehension
- Key and value are separated by a semi-colon

````python
fellowship = ['Samwise', 'Frodo', 'Boromir', 'Gandalf', 'Aragorn', 'Legolas','Merry', 'Pippin', 'Gimli']
fellowship_lengths = {member: len(member) for member in fellowship}
print(fellowship_lengths)
````

# Intro to generator expressions

- Use () instead of [] in a list comprehension and you get a generator object
- Generators -> like list comprehensions, but do not store list in memory
    - Does not construct the list, but is an object that we can iterate over to produce new elements as required
- Looping over a generator produces elements ina list. Can pass to list function to create list, can use "next" to lazily iterate
- Lazy evaluation
    - Evaluation of an expression is delayed until value is needed
- Anything that can be done with a list comprehension (conditionals, filtering) can be done to a generator
### Generator functions
- Define just like any other function, but "yield" instead of "return"

````python
from math import pi, sin

def fracs_of_2pi(n):
    """Generates integer multiples of 2*pi/n"""
    i = 0
    while i <= n:
        yield 2*pi*i/n
        i+=1

sin_dict = {x: sin(x) for x in fracs_of_2pi(10)}

print(list(f"{key:.2f}: {val:.2f}" for key, val in sin_dict.items()))
````
