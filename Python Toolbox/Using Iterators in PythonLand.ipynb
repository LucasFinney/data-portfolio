# Intro to Iterators
- Can iterate over objects like lists, strings, and ranges because they are *iterables*
- Iterable
    - Lists, strings, dictionaries, file connections
    - An object with an associated iter() method
    - Applying iter() to an iterable creates an *iterator*
- Iterator
    - Object with a "next()" method to produce the next value

## Iterating over iterables: next()
- To create an iterator, use the function iter and pass it the iterable 

````python
word = 'Da'
it=iter(word)
next(it)
````

- Print all elements of an iterable using *

````python
word = 'Data'
it = iter(word)
print(*it)
````

- Can only print (*it) once, because there will be no more values to print! The iterator *moves along the iterable* in memory.

## Iterating over dictionaries

- Need to apply ".items()" method to iterate over keys *and values*
    - Looping over a dictionary without .items() will just return the keys, not the key-value pairs

````python
pythonistas = {'hugo': 'bowne-anderson', 'francis': 'castro'}
for value in pythonistas:
    print(value)
for key, value in pythonistas.items():
    print(key,value)
````

# Playing with iterators

## Using enumerate()
- Takes any iterable as an argument, returns an "eunmerate" object -> Pairs of index & values from original iterable

````python
guardians = ['Gamora', 'Starlord', 'Drax', 'Rocket', 'Groot']
g = enumerate(guardians)
g_list = list(g)
print(g_list)
````

- enumerate() itself is an iterable

````python
for index, value in g_list:
    print(index, value)
````

- Enumerate starts the index at 0 by default (as usual for Python), but we can use the start kwarg to change the start index

## Using zip()
- Returns an *iterator of tuples*

````python
first_names = ['Tony', 'Steve', 'Bucky','Wanda']
last_names = ['Stark', 'Rogers', 'Barnes', 'Maximoff']
z =  zip(first_names,last_names)
print(list(z))
````

# Using iterators to load large files into memory
- Too much data from a database or API -> Load in chunks using an iterator
- Using pandas: read_csv() and chunksize argument
Ex.
```
import pandas as pd
result = []
for chunk in pd.read_csv('data.csv',chunksize=1000):
    result.append(sum(chunk['x']))
total = sum(result)
print(total)
```
- Sums everything in column 'x' of each chunk 1000 at a time
