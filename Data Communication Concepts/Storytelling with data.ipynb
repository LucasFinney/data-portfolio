# Fundamentals of storytelling

## Data Storytelling road
 Story -> Tech or not tech -> Data -> Viz -> Present

## Why are stories needed?
- Results have no impact without proper presentation
- Must *convince* change-averse stakeholders

## What is data storytelling?
- Practice of building a narrative around a set of data and its accompanying visualizations to convey the meaning of the data in a powerful and compelling way
- Stories are memorable and add context/value to the data
- Capture audience's attention, facilitate decision making, and drive change

## Data storytelling
- 3-minutes story:
    - How would you convey the information in 3 minutes?
- Big idea:
    - State the unique point of view in one sentence

- Visuals, Data, and Narrative make the story insightful, explanatory, and concise.

### Data
- Results (e.g predictions) and findings (e.g. data analysis).
- Should be relevant, accurate, and reliable, and include actionable insights.

### Narrative
- A good data story has a compelling and easy to understand narrative
- Includes *only* key points needed to drive change
- A compelling narrative revolves around one central insight
- Must understand audience background so we can clarify facts for them
    - Who are they?
    - What do they know?
- Linear sequence -> Every data point builds on each other until the conclusion is reached

### Visuals
- Graphs should be simple, engaging, but not misleading

# Translating technical results

## Tech or non-tech approach?
- Tech knowledge is a continuum
- Data professionals care about their methods, but the audience will care about results and implications

## How technical?
- Ex. May need to explain why low accuracy predictions affect business outcomes to supply chain specialists
    - They don't care about the model or hyper-parameter tuning, they will care about their *own* pain points

## Translating technical results into stories
- Story should be easy to understand and *engaging* to the audience

## Audience awareness
- What is their background, what do they know? Adjust content to audience.
    - Instead of explaining how the model works, the audience should understand the impact and limitations of predictions
- What do they need to understand?
    - Be conversational, anticipate questions
    - Explaining why we chose predictive variables may not be useful, but we should explain the context in which our model works 
- What level of information do they need?
    - Story should help the audience understand what the project is about
    - Listing r values may be overwhelming. Better to explain interaction between customer traits

## ADEPT Technique
- Analogies
    - Instead of explaining how a neural network works, say it learns like a child does by getting feedback on what is right or wrong
- Diagrams
- Examples
- Plain English
- Technical Definitions/Jargon
    - Use acronyms with caution
    - Introduce the full term the first time the acronym is used
    - Limit the use of jargon
    - Include a reference guide or definitions when appropriate

## Focus on impact
- A persuasive story focuses on on impact rather than process
- Instead of "use a non-relational database to make efficient nested queries", focus on "Changing the storage approach will save a lot of time"
- Instead of "Number of rooms shows correlation of 0.7 with house price", focus on "The more rooms in the house, the higher the price"

## Humility
- Not all technical details can be removed
- Be receptive to questions, proactively ensure understanding
- Expect to explain concepts again using different strategies

# Impacting the decision-making process

## Compelling narrative
"A description of connected events that organizes information to engage the audience and make them care for the results or information shared"
- Found insights, now make those meaningful to the target audience
- Prioritize key points to drive change

## Narrative structure
ex. Analyzing sales for a snack company
- Background -> What motivated the analysis? What changed? Who is the focus of the analysis?
    - Ex. Total profit has decreased
- Insight 1 -> What contributed to the problem? Only relevant information
    - Ex. Chip sales up by 20%, sweets down by 30%
- Insight 2 -> Add supporting evidence. Help better explain the cause of problem
    - Ex. Most popular chocolate sales have declined by 50%
- Climax -> Central insight. What would happen if there is no change.
    - Ex. On track to lose $10M next year
- Next steps -> Potential solutions, courses of action. Take proactive steps to guide audience through what to do with our results
    - Ex. Rebrand chocolate, decrease prices, improve ingredients, etc.

## Building narrative
- Change over time
    - Ex. Chocolate sales show a seasonal pattern, lower in summer and higher in winter
- Correlation
    - Ex. Customer rating of chocolate vs. price 
- Comparison
    - Ex. Similarities and differences in chocolate consumption between children and adults
- Clustering
    - Ex. Groups with different coffee and chocolate consumption patterns
