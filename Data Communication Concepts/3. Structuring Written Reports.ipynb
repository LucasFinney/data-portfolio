# Types of reports
- Will discuss: 
    - How to structure
    - Types of reports
    - Reproducibility in DS
    - How to write precise and clear reports

## Written reports
- Explain data analysis project
- Communicate findings
- Conform to industry standrads
- Give recommendations to drive change

## Types of reports
- Informational
    - Factual
    - Short
    - No strict structure
    - Inform about facts

- Analytical
    - Analysis (Relationships/recommendations)
    - Variable length (short or long)
    - Strict structure
    - Drive decisions based on data

## Final Reports
- Elements
    - Data analysis
    - Findings and results
    - Visuals
- Format
    - Usually long
- Audience
    - Need technical details

## Summary reports
- Elements
    - Key findings and recs
    - Visuals
- Format
    - Short (<5 pages)
    - Summary of final report
    - May link to main document
- Audience
    - Decision makers who do not need technical details

## Report structure (Academic/Journal)
1. Introduction
    1. Summarize purpose of report
        - ex. Analysis of product reviews gathered from website, rating predictions based on reviews
    2. Provide contextual information
        - ex. Increase in negative reviews
    3. Present the question of analysis
        - ex. What is causing users to have a bad experience?
2. Body
    1. Data - Description and tables
    2. Methods
        - Ex. NLP and Random Forest
    3. Analysis - Visuals
        - Ex. Graphs with most common words
    4. Results - Description and visuals
        - Ex. 30% negative ratings associated with words "delayed" and "shipping"
3. Conclusions
    1. Restate questions
    2. Summarize important results
    3. Add recommendations for next steps

- This structure is appropriate for academic reports or journals

## Report Structure (Business context)
- 1-3-25 approach
    - 1 page of abstract
    - 3 pages (max) of executive summary
    - 25 pages (max) of details
- Audience
    - Tailor to audience!
        - People with little time will read intro, conclusion, and skim the body
        - Executive team will read intro, conclusions, focus on recommendations
        - Technical stakeholders will focus on the body of the report; understand methods and analysis

# Reproducibility
- Reproducibility - Rerunning the Jupyter Notebook should lead to the same results
- Replicability - Running the project in a different environment (different language, OS) leads to similar results
## Virtues
- Prevents duplication of effort
- Builds upon existing work
- Allows new work to focus on new challenges
- Peer review
- Tool agnostic

## Best Practices
1. Keep track of how results were produced
    - Well documented code
    - List packages and libraries
    - Use version control (git)
2. Avoid manual data manipulation
    - Data versioning
    - Store raw data and intermediate steps
    - Adapt and resolve problems
    - Ex. Imputation -> Impute missing values. Need to keep track which values were replaced
3. Control randomness
    - Random seeds for ML pipelines
    - Control confounding variables
4. Interpretability
    - Understand the cause of a decision or predict results
    - Story with compelling narrative helps stakeholders interpret findings
5. Cite Bibliography
    - Use reference management tools: Endnote, Mendeley, RefWorks
    - Business context is typically less strict, a hyperlink is usually fine

# Write precise and clear reports
- In order to make an impact, others must be able to reproduce results
- Writing must be concise, precise, and avoid confusion

## Empty phrases
- Contain no information ("Well known that..." "Interesting to note that...")

## Concrete nouns
- Avoid "this", "that", or "it"
    - Keeping track of pronoun meaning adds cognitive load, distracts from insights
    - Ex. "This shows an accuracy of 80%" vs. "The model shows an accuracy of 80%..."
- Academic setting prefers passive voice, Business setting prefers active voice

## Redundant adjectives and adverbs
- "Introducing new"... "Done previously"
- Eliminate

## Run-on sentences
- Two or more independent clauses connected incorrectly
    - Ex. "There is a correlation between delayed shipping and customer rating, the shipping delay is the cause for negative review" -> Should be two sentences or connected with dependent clauses (ex. because, so)

- Note: Read "Communicating with Data", Nolan D, Stoudt S.

# Case study: report on credit risk
## Scenario: Credit risk
- Quantifies probability of defaulting
- Loanme bank wants to predict if a customer is likely to default
- Raw data available
- We performed EDA and trained a predictive model -> Ready to present

## Audience
- Non-technical stakeholders

## Data road
1. Story
    - Background: Increase in defaulting percentage. Want to predict which customers had a high probability of default
    - Insight: People with more unemployment periods and lower income tend to default more
    - Climax: Predict which people are likely to default with 95% accuracy
    - Next steps: Run a trial on a control population
2. Approach
    - Translate technical results for non-technical stakeholders
3. Data
    - Audience persona: Financing Department Director. Interested in implementing automated loan rejection system
    - Appropriate data -> Report relationship between income and loan default, prediction of percentage of customer defaults over the next months
    - Appropriate statistics -> Median age and income for default vs. non default group. Percentage change of # of defaults over time
4. Visuals
    - Boxplot with age vs. default condition
    - Lineplot with percentage change in defaulting customers
5. Presentation
    - Who: Financial Department director
    - Why: Important decisions ahead
    - Content: Key findings and recommendations
    - Channel: Send the reults before meeting
    - Medium: Written summary, analytical report
