# Planning an oral presentation
## Plan a presentation
- Note: Read "Better Presentations", Schwabish
- Consider purpose, audience, message, and appopriate structure

## Ex. Sentiment analysis of product reviews
### Purpose
- Informative -> Current number of negative and positive ratings, words associated with negative reviews
- Instructional -> Explain how the model for sentiment analysis was built
- Persuasive -> Convince non-technical stakeholders of particular actions to reduce negative ratings

### Audience
- Who?
    - Technical Colleague
    - Manager
    - Customer
- How big/what context?
    - Small meeting
    - Conference or large meeting
    - Workshop
- Message
    - What is the central message? What is the one thing they need to remember from the presentation?
    - Opening statement: Capture audience's attention. ex. "Negative ratings scare customers away."
    - Central message: One sentence. "Delayed shipping is the main cause of negative reviews"
    - Closing statement: Sums up presentation and strengthens message. 
        - ex. "Decrease in sales, increase in negative reviews because of delayed shipments. Need to fix shipping to address sales"

### Structure
- Introduction
    - Provide background info
    - Get attention
    - Glimpse of presentation content
- Methods, analysis, and model outputs
    - Might skip if audience is non-technical
- Conclusions and takeaways
    - Refers back to the introduction
    - Contains call-to-action statement and/or next steps
    - Remember, audience is likely more to remember beginning and end of talk

### Outline
- Graphs and visuals
- Sections (5 or less, generally). For our example:
    1. Reason for analysis
    2. Exploratory analysis
    3. Sentiment analysis
    4. Conclusions
    5. Follow-up actions
- Keep time in mind!


# Building presentation slides
## From planning to building
- Slides
    - Support story
    - Short, dynamic (better to have many slides with little content than opposite)
    - Slide count is a bad metric
    - Focus on one message per slide
    - Covering too much will increase cognitive load
- Color
    - Everything you present should convey meaning
    - Use fewest number of colors
    - Consider contrast and color deficiency for readability
- Fonts
    - Serif vs. sans-serif is inconclusive, but general rule of thumb is sans-serif for text on screen, serif for printed text
    - Context? Comic sans for kids, not business
    - Slide content should support what is being said
    - Use short text with large fonts
    - Bold or italicize or emphasis
- Text slides
    - Main points to support message
    - Don't dual purpose the slide deck into a written report
    - Include concise, specific headline to highlight main point
    - Break complex slides into smaller points
- Visualization slides
    - Consider using graphs to replace text-dense points
    - Use layering and highlighting to direct audience attention
    - Use a headline (if needed)
    - One or two full-size graphs per slide
        - One message per slide, avoid overcrowding

# Delivering the presentation

## Practice


### Prepare
- Write a script, but don't memorize
    - Use as a guide
- Become familiar with content
- Anticipate follow-up content

### Rehearsal
- Stand up
- Use the slides
- Speak out loud
- Detect distracting patterns
- Find linking statements
- Anticipate QA

### Delivery
- Be aware of emotions
    - Confidence vs. unsure
- Attention span varies, but expect 5-20 minutes
    - Keep engagement with interactivity, eye contact
    - Talk to audience (not *at* them)
    - Develop a relationship (avoid phrases like "as we know", "obviously")
- Use allocated time (don't go over)
- Pace
    - Pauses are important for understanding
    - Give audience time to think
- Open up for questions during or at the end of the presentation

## An effective oral presentation
- Talk *to* the audience
- Less is more
- Consistent and persistent take-home message
- Practice
- Prepare for follow-up questions

# Avoiding common errors
- Plan, prepare, deliver

## Length
- Many presentations are ineffective because they are too long

## Purpose
- Always state the purpose at the *beginning* of the introduction

## Guiding audience
- Keep audience's attention
- Do not leave all findings to the end, include them throughout the presentation

## Audience involvement
- Engage and involve audience
- Introduce yourself!
- State key assumptions
- Ask questions *of the audience* (know answer, serves as hook)
- Reiterate and relate to the main idea

## Body language
- Emphasis by natural gesture, movements, body language

## Voice tonality
- Use different voice tonalities
    - Speed
        - Fast: urgency, excitement, emotion
        - Slow: importance, introduce new ideas
    - Volume
        - Live: project volume
        - Online: make sure your mic works
    - Intonation
        - Sound interested!
