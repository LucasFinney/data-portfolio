# Selecting the right data

*"Selecting the right data implies including enough contextual insights in a story to better support the main point without overloading the audience with information"*
- Use minimal amount of data to support our story

- Data is a key part of a data narrative (duh)
- Garbage in, garbage out -> Bad data, poor quality insights and results
- Must adapt data to audience

## Stakeholders
- Defined as any person interested in the project outcome or a decision or activity derived from it
    - Can be technical or non-technical

## Identifying personas
- Description, interests, knowledge
- Helps to select tailored findings

ex. The communicatb food project personas
- The Executive Team
    - Role: Executive level (CEO, investor, director, founder)
    - Knowledge: Fundamentals
    - Interest: Inform their decisions based on findings
- Project Manager
    - Interest: Project aligns with company objectives
    - Right data: 
        - Summary data -> $2M cost of marketing campaign
        - Metrics: 10% monthly increase in number of customers, 2% risk of declining profits
- Tech team
    - Role: Project collaborator, Technical supervisor, 
    - Knowledge: Expert (technical aspects)
    - Interest:
        - Replicate or continue project
- General audience
    - Role: Customer or other department staff
    - Knowledge: Novice or generalist
    - Interests: To understand the general results and impact of the project
    - Right data: Historical data on decline in profits, correlation/impact (chocolate needs rebranding)


## Audience skepticism
- Audience may be more or less receptive to results
- Be prepared to bring in more data/insights/info to convince skeptics
- First convince yourself, then a friend, then a skeptic

# Showing relevant statistics
## Variations of data
- Absolute (ex. 10,000 more units sold) -> Absolute change is often okay when focusing on one variable
- Relative (ex. 20% growth YoY)
- Ratio (ex. Revenue per customer) -> Helps to normalize values

## Aggregates
- Representative value
    - Totals/Counts
    - Mean
    - Median
- Mean can be misleading (due to outliers)

# Visualizations for different audiences
- Adjust visualization to suit the message we want to convey
- Consider audience expertise and familiarity

## Tailored message (chocolate project example)
- Need to communicate one message to an investor, one to a technical lead
- Investor
    - Marketing campaign to avoid 20% revenue drop
    - Show a line graph with forecasts (launch vs. no launch)
    - Show influence of different factors on customer behavior
- Tech lead
    - Model predeicted revenue decline (RMSE: 1.2)
    - Show how well the model performed on historical data
    - Show feature importance in model

## More best practices
- Pareto principle (80-20 rule) -> Aggregate data to reduce noise and distil insights
- Approachable and engaging visuals
- How many insights from visualization, and how quickly do they get them?
- Less is more

## McCandless method
1. Introduce visualization by name
    - Graph headline
    - Clear and obvious (ex. "Chocolate sales have been decreasing since 2017)
    - Use y vs x technique *WHEN IN DOUBT*
2. Anticipate audience's questions
    - Focus on story, not on decoding graph
3. State insights
    - Explain what the audience is seeing!
4. Help the audience relate
    - Why do the insights matter?
    - How can they be acted upon?

# Choosing the appropriate format
- Data storytelling road: Design story -> Define approach -> Select data -> Choose visualization -> Present data
- Effective formats show key information in an engaging and easy to understand way
## Main formats
- Written reports
- Oral presentations

## Presentation Strategy
- Conside audience, content, requirements, channel to determine presentation format 

## Stakeholders
- **Who** are we presenting to?
- **Why** do they need to know about our findings? 
    - Accountability
    - Methodology?
- **How** are they going to use our findings?
    - Make decisions
    - Start a new project
- **What** information do they need?
    - Results
    - Impact

## Requirements
- Time:
    - Do they have time to read a report, or should it be a short meeting?
- Authority:
    - Do they report to someone else, need a document to back up a claim?
- Time zone:
    - Are they in a different timezone, so written communication is preferred?

## Consumption
- How will the information be consumed?
- Format
    - Document
    - Jupyter notebook
    - Blog post/articel
    - Slide deck
- Delivery
    - Live
    - Asynchronous
- Audience Size
    - Conference room vs. ball room!

## Oral communication
- Advantages
    - Relationship with audience
    - Immediate feedback
    - Non-verbal cues
- Disadvantages
    - No permanent record
    - Not suitable for long messages

## Written communication
- Advantages
    - Permanent record
    - Shared easily with a large audience
    - Less emotional reaction
    - Suitable to share code with colleagues
- Disadvantages
    - Harder to adapt to audience
    - No immediate feedback

## Example: Presenting findings to communicatb CEO
- Who? CEO
- Why? Monthly update
- What? Accountability
- Content: Conclusions
- Requirement: Time
- Approach -> Oral Presentation
