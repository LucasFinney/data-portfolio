```python
%matplotlib inline
# Exercise: What genres are most likely to have sequels?

# Import pandas
import pandas as pd

# Import some of the course datasets 
movies = pd.read_pickle("datasets/movies.p")
genres = pd.read_pickle("datasets/movie_to_genres.p")
sequels = pd.read_pickle("datasets/sequels.p")

# the "movies" and sequels dataset has an "id" column rather than "movie_id", so that needs to change
genres.rename(columns={"movie_id":"id"},inplace=True)

# Just get movies with sequels from the sequel set
sequels.dropna(inplace=True)

# merge movies with genres 
movies_genres_w_sequels = movies.merge(genres,on="id").merge(sequels,on="id")

import matplotlib.pyplot as plt
movies_genres_w_sequels['genre'].value_counts().plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Genre Frequency in Movies with Sequels')
plt.tight_layout()
plt.show()
```
