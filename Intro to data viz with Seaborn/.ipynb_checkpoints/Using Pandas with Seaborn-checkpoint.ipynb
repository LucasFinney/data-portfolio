```python
# Importing the course packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importing the course datasets
country_data = pd.read_csv('../datasets/countries-of-the-world.csv', decimal=",")
mpg = pd.read_csv('../datasets/mpg.csv')
student_data = pd.read_csv('../datasets/student-alcohol-consumption.csv', index_col=0)
survey = pd.read_csv('../datasets/young-people-survey-responses.csv', index_col=0)
```

```python
# Cleaning Data
df_raw = pd.read_csv("../datasets/raw-responses.csv")
df_raw = df_raw.rename(columns = {"Unnamed: 0":"participant_id", "q0001":"how_masculine","q0002":"how_important", "age3":"age"})
df = df_raw.loc[:,["participant_id","age","how_masculine","how_important"]]
df["how_masculine"] = df["how_masculine"].str.replace("masculine",repl="")
df["how_important"] = df["how_important"].str.replace("important",repl="").str.replace("too", repl="very")
df.to_csv("../datasets/masculinity.csv")
```

# Using pandas with Seaborn

```python
df = pd.read_csv("../datasets/masculinity.csv", index_col = 0)
```

```python
sns.countplot(x="how_important",data=df) # or sns.countplot(x=df["how_masculine"]) equivalently
plt.show()
```

# Adding a third variable with hue

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="sex", data=tips)
plt.show()
```

### Adding a 4th variable with size!

```python
sns.scatterplot(x="total_bill",y="tip",hue="sex", data=tips,size="size")
plt.show()
```

## Using hue with count plots

```python
sns.countplot(x="how_masculine", hue="how_important", data=df)
plt.show()
```

```python
sns.countplot(x="sex",hue="smoker",data=tips)
plt.show()
```
