```python
%matplotlib inline
# Importing the course packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importing the course datasets
country_data = pd.read_csv('../datasets/countries-of-the-world.csv', decimal=",")
mpg = pd.read_csv('../datasets/mpg.csv')
student_data = pd.read_csv('../datasets/student-alcohol-consumption.csv', index_col=0)
survey = pd.read_csv('../datasets/young-people-survey-responses.csv', index_col=0)
masculinity_data = pd.read_csv('../datasets/masculinity.csv', index_col=0)
```

# Changing plot style and color

## Changing figure style
- Default figure style is white
- Can set value to "whitegrid", "whitetick", "dark", "darkgrid", etc.

```python
sns.set_style('whitegrid')
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=tips, kind="violin", hue="day", legend=False, dodge=False)
plt.show()
```

## Changing the palette
- Changes color of elements
- Diverging palettes (RdBu, PRGn, RdBu_r) -> Red to blue, purple to green, etc. _r will reverse the palette
- Sequential palettes give a discrete gradient along shades of grey, blues, etc.

```python
sns.set_palette("RdBu_r")
category_order = ["No answer", "Not at all ", "Not very ", "Somewhat ", "Very "]
sns.catplot(x="how_masculine", data=masculinity_data, order=category_order, kind="count", hue="how_masculine")
plt.show()
```

## Changing the scale
- Use sns.set_context()
- Settings are "paper", "notebook", "talk", "poster"

## Creating informative visualizations

### FacetGrid vs. AxesSubplot objects
- FacetGrid consists of one or more AxesSubplots
- FacetGrid
    - relplot(), catplot(), can create subplots
- AxesSubplot
    - scatterplot(), countplot(), etc. Only creates a single plot

### Adding a title to FacetGrids
- Save the plot to a variable, then call figure.suptitle()

### Titles for subplots
(where g is the catplot object)
- g.set_titles("This is {col_name}")

### Axis labels
- g.set(xlabel=..., ylabel=...)
- Rotate labels with plt.xticks(rotation=...)
