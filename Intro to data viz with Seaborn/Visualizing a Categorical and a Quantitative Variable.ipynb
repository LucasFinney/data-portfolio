```python
%matplotlib inline
# Importing the course packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importing the course datasets
country_data = pd.read_csv('../datasets/countries-of-the-world.csv', decimal=",")
mpg = pd.read_csv('../datasets/mpg.csv')
student_data = pd.read_csv('../datasets/student-alcohol-consumption.csv', index_col=0)
survey = pd.read_csv('../datasets/young-people-survey-responses.csv', index_col=0)
masculinity_data = pd.read_csv('../datasets/masculinity.csv', index_col=0)
```

# Count plots and bar plots

## catplot()
- Allows multiple types of categorical plots, similar to relplot()
- Change order using "order=" kwarg

```python
category_order = ["No answer", "Not at all ", "Not very ", "Somewhat ", "Very "]
sns.catplot(x="how_masculine", data=masculinity_data, hue="how_masculine", kind="count", order=category_order)
plt.show()
```

## Bar plots
- Quantitative value as a function of categorical
- Use "errorbar" parameter to control display (ci vs. sd vs. none)

```python
# Import tips dataset
tips = sns.load_dataset("tips")

# Create a bar plot of total_bill by day
sns.catplot(x="day", y="total_bill", data=tips, kind="bar", hue="day", legend=False, dodge=False)
plt.show()
```

# Box plots
- Specify kind = "box" in catplot()
- Create subplots using row and col parameters
- Omit outliers with showfliers=False
- Change percentiles of whiskers using whis=[lower, upper] (or set length to a multiplier of the IQR by writing whis=x)

```python
g = sns.catplot(x="time", y="total_bill", data=tips, kind="box", hue="time", dodge=False, legend=False, showfliers=False,whis=[0, 100])
plt.show()
```

# Point plots
- Show the mean of a quantitative variable organized by categorical variables
- Vertical lines show 95% CI (default)
- Similar to bar plots, but it's easier to compare the heights of subgroup points
- Control estimator using "estimator =... " argument
- Add "caps" argument to adjust display of ends of interval

```python
sns.catplot(x="smoker", y="total_bill", data=tips, kind="point", capsize=0.2)
plt.show()
```
