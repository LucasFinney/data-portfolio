# User-defined functions

# Multiple Parameters and Return Values

```python
# Example:
def raise_to_powern(x,n):
    """Raise x to the power of n."""
    new_value = x ** n
    return new_value
print(raise_to_powern(2,3))
```

## Return multiple values


### A quick jump into tuples
- Can contain multiple values (like a list)
- Immutable - can't modify values
- Constructed using parentheses ()

### Unpacking tuples
- Can access individual elements in a tuple just like in a list (using square brackets)

```python
even_nums = (2,4,6)
a,b,c = even_nums
print(f"a: {a}, b: {b}, c: {c}")
```

### Multi-return example

```python
def raise_both(x,y):
    """ Raise x to the power of y and y to the power of x."""
    x_raised = x ** y
    y_raised = y ** x
    tuple = (x_raised, y_raised)
    return tuple

a,b = (4,5)
a2, b2 = raise_both(a,b)
print(f"{a} raised to the power of {b} is {a2}")
print(f"{b} raised to the power of {a} is {b2}")
```
