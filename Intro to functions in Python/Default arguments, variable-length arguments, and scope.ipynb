# Scope in Functions
- Defines where in a program an object or name may be accessible
- Three Types
    - Global scope - defined in the main body of a script
    - Local scope - defined inside a function
    - Built-in scope - names in the built-ins module
- If a variable is defined both locally and globally, Python looks locally first
- To alter a value *in a function call*, use the global keyword

```python
from random import randint

x = 10

def fuck_my_shit_up(value):
    """ adds a random number to the input value """
    global x
    x = value + randint(1,10)
    return x

print(x)
print(fuck_my_shit_up(x))
print(x)
```

# Nested functions
Why?
- Repeated processes

```python
# Nested function example

def mod2plus5(x1,x2,x3):
    """Returns the remainder plus 5 of three values"""
    def inner(x):
        return x % 2 + 5
    return inner(x1), inner(x2), inner(x3)

print(mod2plus5(1, 2, 3))
```

```python
def make_character(str, dex, name):
    """ Create a character """
    def print_info():
        print(f"Name: {name}")
        print(f"Strength: {str}")
        print(f"Dexterity: {dex}")
    return print_info

alice = make_character(10, 5, "Alice")
alice()
```

# Default and flexible arguments

```python
# ex. default power
def power(number, pow=1):
    """Raise number to the power of pow"""
    new_value = number ** pow
    return new_value

power(2)
```

```python
# ex. Flexible arguments (*args) -> Use when you don't know how many arguments will be passed to the function
def flexible_args(*args):
    # *args means that all arguments will be passed as a tuple, called args
    total = sum(args)
    return total

print(flexible_args(1, 2, 3, 4, 5))
```

```python
# ex. Flexible keyword arguments (**kwargs) -> Use when you don't know how many keyword arguments will be passed to the function
def flexible_kwargs(**kwargs):
    # **kwargs means that all keyword arguments will be passed as a dictionary, called kwargs
    for key, value in kwargs.items():
        print(f"{key}: {value}")

flexible_kwargs(name="Alice", age=30, city="New York")
```
