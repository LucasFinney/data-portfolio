# Lambda Functions
- "quicker way to write functions on the fly"

````python
# Example: "raise to power"
raise_to_power= lambda x, y: x ** y
print(raise_to_power(2, 3))
````

- Example use case: the map function: ```map(func,seq)```
    - Takes two arguments, a function and a sequence
    - Applies that function over all elements in the sequence

````python
from random import random

# Randomly capitalize letters in a string using a lambda function and map
string = "hello world"
random_caps = map(lambda x: x.upper() if random() < 0.5 else x.lower(), string)
random_caps_string = ''.join(random_caps)
print(random_caps_string)
````

# Error Handling

- Errors caught in execution are "exceptions"
- Catch exceptions with try-except clause

````python
# Ex. Square Root function
def sqrt(x):
    """Returns the square root of a number"""
    if x<0:
        raise ValueError('x must be non-negative')
    try:
        return x ** 0.5
    except TypeError:
        print('x must be an int or float')

sqrt(-1)
````
