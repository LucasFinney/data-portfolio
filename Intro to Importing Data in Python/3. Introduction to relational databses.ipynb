# Intro to relational databases

## What is a relational database?
- Based on relational model of data
- "Table" is equivalent to a dataframe -> Columns for attributes/variables, one row per "instance" or measurement of that variable (ex. D&D stats are columns, one row per character)
    - Each row has a unique identifier -> "Primary key"
- Databases consist of multiple linked tables -> link by keys to other tables
- Ex. Northwind Database -> Synthetic database of sales for a  fictional company
    - 3 tables: orders, customers, employees
    - Link orders to employees and customers using CustomerID and EmployeeID
- Codified in Codd's 12 rules/commandments

## Relational database management systems
- PostgreSQL (or Postgres)
- MySQL
- SQLite
- SQL = Structured Query Language -> how you communicate with a database to access or update data
