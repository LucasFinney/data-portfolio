## Flat Files

- Text files containing records (table data without structured relationships)
- Record: row of fields or attributes
- Attribute: A variable
- Headers: Row describing the attributes or features of each column below

# Importing flat files using NumPy

- `np.loadtxt()` and `np.genfromtxt()` are of interest for creating NumPy arrays from text
- In `.loadtxt()`, 
    - use `delimiter=` to specify delimiter
    - `skiprows=` to skip rows (ex. skip the header) 
    - `usecols=[list]` to list indices of columns to keep (ex. [0,2] for cols 1 and 3)
    - `dtype =` to specify datatype of all entries

```python
import numpy as np
filename = 'datasets/mnist_kaggle_some_rows.csv'
data = np.loadtxt(filename, delimiter=',') # Default delimiter is white space
data
```
