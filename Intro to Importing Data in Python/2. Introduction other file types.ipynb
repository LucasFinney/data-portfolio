# Intro to other file types

- Will use Excel spreadsheets, MATLAB, SAS, Stata, and HDF5

## Pickled files (`.pkl`)

- Native to Python
- Serialized  (converted to bytestream)
- Importing is similar to importing raw text in Python
- import arg `'rb'` specifies 'read only' and 'binary'

## Excel spreadsheets
- Many ways, but can use `data = pd.ExcelFile(file)`
- Find the sheets using `data.sheet_names`
- Convert to a dataframe using `df = data.parse(sheet_name)` with sheet_name as a string or `df=data.parse(index)` with sheet index as a float
- `data.parse(sheet_name, usecols= , skiprows = , names = [])`
    - Use `skiprows` as usual, `names` to specify the names of the columns with a list (one per column), `usecols = ` to specify which columns to keep by index(?)

## Exploring the current working directory

- Use the `os` library -> `import os`
- Use `wd = os.getcwd()` to save the current working directory to the variable `wd`
- Use `os.listdir(wd)` to list the contents of the directory

```python
import os
wd = os.getcwd()
os.listdir(wd)
```

# Importing SAS/Stata files using pandas

- SAS -> Commmon in business analytics and biostatistics
- Stata -> academic social sciences research (economics, epidemiology)

## Importing SAS files
- Most commonly have extension `.sas7bdat` or `.SAS7BDAT`
- sas7bdat library must installed locally (just like pandas or whatever)
- Use a context manager (ie. `with SAS7BDAT(location) as file:`)
- `from` lower_case `import` UPPER_CASE -> `from sas7bdat import SAS7BDAT`

```python
import pandas as pd
from sas7bdat import SAS7BDAT
with SAS7BDAT('./datasets/sales.sas7bdat') as file:
    df_sas = file.to_data_frame()

df_sas.head()
```

## Importing Stata files

- No context manager needed
- Function built into pandas `pd.read_stata()`

```python
import pandas as pd
data = pd.read_stata('datasets/disarea.dta')
data.head()
```

# Importing HDF5 files

- HDF5 -> Hierarchical Data Format version 5
- Standard for storing large quantities of numerical data
- Datasets can be hundreds of gigs or terabytes
- Could be up to *exabytes*

- Need `h5py` package!!!
    - Use `data = h5py.File(filename, 'r')` to read h5py into data variable ('r' is to read)

```python
import h5py
data = h5py.File('datasets/L-L1_LOSC_4_V1-1126259446-32.hdf5', 'r')
print(type(data))
```

## Structure of HDF5 files
- Explore like a dictionary using method `.keys()`

```python
for key in data.keys():
    print(key)
```

- Each `key` is a 'directory'
    - `meta` contains meta-data for the file
    - `quality` refers to data quality
    - `strain` refers to strain data from interferometer
- Find values of metadata like above

```python
for key in data['meta'].keys():
    print(key)
```

- Maybe we want the description and detector?
    - Convert values to numpy arrays -> `np.array(data['directory']['value'])`

```python
import numpy as np
description, detector = np.array(data['meta']['Description']), np.array(data['meta']['Detector'])
print(description, detector)
```

```python
print(np.array(data['meta']['Description']), np.array(data['meta']['Detector']))
```

# Importing MATLAB files

- `.mat` files
- Use standard `SciPy` library:
    - `scipy.io.loadmat()` - read
    - `scipy.io.savemat()` - write
- `.loadmat()` imports data as a dictionary
    - Keys are variable names
    - Values are objects assigned to variables

```python
import scipy.io
filename = 'datasets/ja_data2.mat'
mat = scipy.io.loadmat(filename)
print(mat.keys())
```
